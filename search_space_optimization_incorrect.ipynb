```python
ROBOT_SPEED = 1

test_runs = 100000
total_distance_random = 0

curr_position = np.array([[0.25, 0.9277887344360352,-0.5]])

positions = np.array([[1.2,1.87,-1.4],
                      [0.8,1.87,-1.4],
                      [0.3,1.67,-1.7]])

for i in range(test_runs):    
    locations  = [0,1,2]
    population = [0.7,0.20,0.10]
    true_placement = np.random.choice(locations, p=population)
    
    random = [1/3,1/3,1/3]
    random_placement = np.random.choice(locations, p=random)
    random_placement_idx = locations.index(random_placement)
    
    
    if(random_placement==true_placement):
        total_distance_random += np.linalg.norm(positions[random_placement]-curr_position)
        curr_position = positions[random_placement]
    
    else:
        total_distance_random += np.linalg.norm(positions[random_placement]-curr_position)
        curr_position = positions[random_placement]

        while(random_placement!=true_placement):
            locations.pop(random_placement_idx)
            random.pop(random_placement_idx)
            
            total = 0
            for i in range(len(random)):
                total += random[i]
            
            for i, element in enumerate(random):
                random[i] = element/total
            
            random_placement = np.random.choice(locations, p=random)
            random_placement_idx = locations.index(random_placement)
            
            if(random_placement==true_placement):
                total_distance_random += np.linalg.norm(positions[random_placement]-curr_position)
                curr_position = positions[random_placement]
                break 
            
            else:
                total_distance_random += np.linalg.norm(positions[random_placement]-curr_position)
                curr_position = positions[random_placement]

print(total_distance_random)
print("avg time to find object=",total_distance_random/ROBOT_SPEED/test_runs)
```
